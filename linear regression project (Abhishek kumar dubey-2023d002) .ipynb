```python
import pandas as pd 
import numpy as np 
```

```python
Houseing=pd.read_csv("C://Users//Lenovo//Desktop//Housing.csv")
```

```python
H=Houseing
```

```python
H
```

```python
H.head()
```

```python
x=H[["area","bedrooms","stories","parking","bathrooms"]].copy()
y=H[["price"]].copy()
```

```python
x
```

```python
y
```

```python
 x["intercept"] = 1
```

```python
x = x[["intercept","area","bedrooms","stories","parking","bathrooms"]]
```

```python
x
```

```python
x_T = x.T
```

```python
x_T
```

```python
B = np.linalg.inv(x_T @ x) @ x_T @ y
```

```python
B
```

```python
B.index = x.columns
```

```python
B
```

```python
predictions = x @ B
```

```python
predictions
```

```python
SSR = ((y - predictions) **2).sum()
```

```python
SST = ((y-y.mean())**2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2 = 1 - (SSR/SST)
```

```python
R2
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(H[["area","bedrooms","stories","parking","bathrooms"]], H[["price"]])
```

```python
lr.intercept_
```

```python
lr.coef_
```
